```python
import pandas as pd

from source.utils.augmentations import get_albumentation_augs
from source.datasets.ssl_dataset import SSLDataset
from source.utils.general import read_config


config = read_config("../config.yml")
config.dataset.image_folder = "../test_data/images/"
config.dataset.csv_path = "../test_data/test.csv"
df = pd.read_csv(config.dataset.csv_path)
dataset = SSLDataset(config, df, get_albumentation_augs(config)["train"])
```

```python
import torch
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

MEAN = torch.as_tensor([0.485, 0.456, 0.406]).view(3, 1, 1)
STD = torch.as_tensor([0.229, 0.224, 0.225]).view(3, 1, 1)


for data_point in dataset:
    view1 = ((data_point["view1"] + MEAN) * STD).permute(1, 2, 0).numpy()
    view2 = ((data_point["view2"] + MEAN) * STD).permute(1, 2, 0).numpy()
    canvas = np.hstack((view1, view2))

    plt.axis("off")
    plt.imshow(canvas)
    plt.show()
```
